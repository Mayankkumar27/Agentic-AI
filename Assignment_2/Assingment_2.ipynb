```python
import wikipedia
from duckduckgo_search import DDGS

def web_search(query):
    with DDGS() as ddgs:
        results = list(ddgs.text(query, max_results=5))
        output = ""
        for r in results:
            output += f"- {r['title']}: {r['body']}\n"
        return output

def wiki_search(query):
    try:
        summary = wikipedia.summary(query, sentences=10)
        return summary
    except:
        return "Wikipedia result not found."

print("✅ Tools ready")
```

```python
from groq import Groq

client = Groq()

def research_agent(topic):
    print(f"🔍 Web search 1: {topic}")
    web1 = web_search(topic)

    print(f"📖 Wikipedia: {topic}")
    wiki = wiki_search(topic)

    print(f"🔍 Web search 2: {topic} latest trends 2025")
    web2 = web_search(topic + " latest trends 2025")

    print(f"🔍 Web search 3: {topic} challenges problems")
    web3 = web_search(topic + " challenges problems")

    combined = f"""
    WEB SEARCH RESULTS:
    {web1}

    WIKIPEDIA:
    {wiki}

    LATEST TRENDS:
    {web2}

    CHALLENGES:
    {web3}
    """
    print("\n✅ Research done! Summarizing...\n")

    response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                        {"role": "system", "content": "You are an expert research analyst. Summarize research data into a comprehensive structured summary."},
                        {"role": "user", "content": f"Topic: {topic}\n\nResearch Data:\n{combined}\n\nProvide a detailed summary covering: overview, key facts, statistics, trends, challenges, and future outlook."}
                    ],
            max_tokens=2000
        )
    return response.choices[0].message.content

print("✅ Research agent ready")
```

```python
from datetime import date

def generate_report(topic, research_summary):
    response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                        {"role": "system", "content": "You are a professional report writer. Write detailed, formal research reports."},
                        {"role": "user", "content": f"""
                                            Write a complete research report on: {topic}

                                            Research Summary:
                                            {research_summary}

                                            MUST include all these sections exactly:

                                            ================================================
                                            COVER PAGE
                                            ================================================
                                            Title: Research Report on {topic}
                                            Date: {date.today().strftime("%B %d, %Y")}
                                            Prepared by: Autonomous Research Agent (Groq LLaMA3 + DuckDuckGo + Wikipedia)

                                            ================================================
                                            1. INTRODUCTION
                                            ================================================
                                            (Write 150+ words: background and purpose)

                                            ================================================
                                            2. KEY FINDINGS
                                            ================================================
                                            (Write key facts and statistics as bullet points)

                                            ================================================
                                            3. CHALLENGES
                                            ================================================
                                            (Write current problems and ethical concerns)

                                            ================================================
                                            4. FUTURE SCOPE
                                            ================================================
                                            (Write emerging trends and opportunities)

                                            ================================================
                                            5. CONCLUSION
                                            ================================================
                                            (Write final summary and recommendations)

                                            Each section must be at least 150 words. Be detailed and professional.
                                            """}
                    ],
            max_tokens=3000
        )
    return response.choices[0].message.content

print("✅ Report generator ready")
```

```python
# ✏️ CHANGE TOPIC HERE FOR SAMPLE 2
TOPIC = "Impact of AI in Healthcare"

print(f"\n{'='*60}")
print(f"  AUTONOMOUS RESEARCH AGENT")
print(f"  Topic: {TOPIC}")
print(f"{'='*60}\n")

research_summary = research_agent(TOPIC)
print("\nRESEARCH SUMMARY:\n")
print(research_summary)
```

```python
print("\nGenerating final report...\n")
report = generate_report(TOPIC, research_summary)

import os
os.makedirs("outputs", exist_ok=True)
safe_name = TOPIC.replace(" ", "_").lower()[:40]
filename = f"outputs/report_{safe_name}.txt"

with open(filename, "w", encoding="utf-8") as f:
    f.write(report)

    print(f"✅ Saved: {filename}\n")
    print("="*60)
    print(report)
```
